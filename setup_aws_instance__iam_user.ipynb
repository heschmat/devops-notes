```python
"""
🔐 1. PEM File (Private Key)

A PEM file (.pem) is a private key file used to securely SSH into a Linux EC2 instance.

▶️ Purpose:
- Used for SSH access to the EC2 instance.
- It's part of a key pair (public + private):
    - The public key is stored on the EC2 instance.
    - The private key (PEM file) is kept by you, and used with SSH.


N.B. You download the pem file once, when creating the EC2 key pair. Don't lose it — AWS doesn't let you download it again.
If lost, you'll need to create a new key pair and manually update the instance (if possible).
""";
```

```python
"""
🔑 2. Access Key (Access Key ID + Secret Access Key)

▶️ What it is:
A pair of credentials used by AWS IAM users to programmatically access AWS services via:
- CLI
- SDKs
- Terraform, Boto3 ...

▶️ Purpose:
To authenticate to AWS APIs (not to SSH into instances).

Examples:
List s3 buckets

$ aws s3 ls --region us-east-1

""";
```

```python
# now should see this in the `security_credentials` section of the user profile.

# you can also create a key from this section as well.
```
